```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(X)")
plt.show()
```

```python
x = 0
lr = 0.1
```

```python
plt.plot(X,Y)
for i in range(50):
    if i%10 == 0:
        print(f"i is {i}, x value is {x}")
    grad = 2*x - 10
    x = x - lr*grad
    plt.scatter(x,(x-5)**2)

print(f"finalx value is {x}")
```

As we can see the x value is coming to 5, which is the global minima.
